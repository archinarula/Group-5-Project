```python
import pandas as pd
import datetime as dt
```

```python
file1 = "Resources/cases-brazil-cities-time.csv"
file2 = "Resources/NY_data-by-day.csv"
```

```python
brazil_df = pd.read_csv(file1)
brazil_df
```

```python
brazil_df['city'].unique()
```

```python
SaoPaulo_df = brazil_df[brazil_df['city']=='São Paulo/SP']
SaoPaulo_df
```

```python
col_nm = SaoPaulo_df.columns.tolist()
col_nm = col_nm[:3]+col_nm[4:5]+col_nm[10:12]+col_nm[3:4]+col_nm[5:10]+col_nm[12:]
col_nm
```

```python
SaoPaulo_df = SaoPaulo_df[col_nm]
SaoPaulo_df
```

```python
SaoPaulo_df = SaoPaulo_df.replace(to_replace=r'^São.*', value='Sao_Paulo', regex=True)
SaoPaulo_df
```

```python
oppath = "Output/SaoPaulo.csv"
SaoPaulo_df.to_csv(oppath)
```
