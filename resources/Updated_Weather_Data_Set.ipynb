```python
import pandas as pd
import numpy as np
from datetime import date
```

```python
df= pd.read_csv('fa037ddb81b7f7f0a0d1a0ebd131858e.csv')
df.head()
```

```python
df.columns
```

```python
df.drop(df.columns.difference(['dt','dt_iso','city_name','lat','lon','temp','temp_min','temp_max','humidity']), 1, inplace=True)
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['dt_iso'].str[:10])
```

```python
df['time'] = df['dt_iso'].str[10:13]
df
```

```python
df[['city_name']].value_counts()
```

```python
df_agg = df.groupby(['date','city_name','lat','lon']).\
               agg(temp_mean = pd.NamedAgg(column='temp',aggfunc='mean'),
                  humidity_mean = pd.NamedAgg(column='humidity',aggfunc='mean'),
                  temp_min = pd.NamedAgg(column='temp_min',aggfunc='min'),
                  temp_max = pd.NamedAgg(column='temp_max',aggfunc='max') 
                  ).reset_index()
```

```python
df_agg = df_agg.sort_values(['city_name','date'],ascending=True)
df_agg
```

```python
df_agg['temp_mean']=df_agg['temp_mean'].round(decimals=2)
df_agg['humidity_mean']=df_agg['humidity_mean'].round(decimals=2)
df_agg
```

```python
df_agg = df_agg.sort_values(['city_name','date'],ascending=True).reset_index()
```

```python
df_agg['1_7_day_avg_temp']= (df_agg.groupby(['city_name'],sort=False)['temp_mean']
                                   .transform(lambda x: x.shift(1).rolling(window=7).mean())
                                   .reset_index(drop=True))

df_agg['8_14_day_avg_temp']= (df_agg.groupby(['city_name'],sort=False)['temp_mean']
                                   .transform(lambda x: x.shift(8).rolling(window=7).mean())
                                   .reset_index(drop=True))
```

```python
df_agg['1_7_day_avg_humidity']= (df_agg.groupby(['city_name'],sort=False)['humidity_mean']
                                   .transform(lambda x: x.shift(1).rolling(window=7).mean())
                                   .reset_index(drop=True))

df_agg['8_14_day_avg_humidity']= (df_agg.groupby(['city_name'],sort=False)['humidity_mean']
                                   .transform(lambda x: x.shift(8).rolling(window=7).mean())
                                   .reset_index(drop=True))
```

```python
df_agg
```

```python
df_agg['1_7_day_avg_temp']=df_agg['1_7_day_avg_temp'].round(decimals=2)
df_agg['8_14_day_avg_temp']=df_agg['8_14_day_avg_temp'].round(decimals=2)
df_agg['1_7_day_avg_humidity']=df_agg['1_7_day_avg_humidity'].round(decimals=2)
df_agg['8_14_day_avg_humidity']=df_agg['8_14_day_avg_humidity'].round(decimals=2)
df_agg
```

```python
df_agg['percent_change_in_temp']=((df_agg['8_14_day_avg_temp']-df_agg['1_7_day_avg_temp'])/df_agg['8_14_day_avg_temp'])*100
```

```python
df_agg['percent_change_in_humidity']=((df_agg['8_14_day_avg_humidity']-df_agg['1_7_day_avg_humidity'])/df_agg['8_14_day_avg_humidity'])*100
```

```python
df_agg
```

```python
df_agg['percent_change_in_temp']=df_agg['percent_change_in_temp'].round(decimals=2)
df_agg['percent_change_in_humidity']=df_agg['percent_change_in_humidity'].round(decimals=2)
```

```python
df_agg
```

```python
df_agg.to_csv('NYC_SaoPaulo_Weather.csv',index=False)
```
