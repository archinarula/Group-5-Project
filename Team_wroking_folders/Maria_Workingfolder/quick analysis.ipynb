```python
df_sp_data = pd.read_csv('Resources/SaoPaulo_Dataset2.csv', parse_dates=['date'])
df_sp_data.head(5)
```

```python
# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
df_sp_data.plot(figsize=(28, 10), kind='line')
plt.ylabel('Num of Cases', fontsize='20')
plt.xlabel('Date')
plt.ylim(0, 905500)
plt.yticks(fontsize='18')
plt.xticks(fontsize='18')
plt.title('COVID-19 by City Type', fontsize='30')

plt.savefig("Analysis/SP.png")
```
