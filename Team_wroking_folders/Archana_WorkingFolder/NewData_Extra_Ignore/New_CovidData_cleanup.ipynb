# Covid new cases data cleanup-Sao Paulo and New York

```python
# Import Libraries Required
import pandas as pd
import numpy as np
import os
from datetime import datetime
```

```python
# Add the Resources path variables:
NY_data = "NY_new_covidcase.csv"
SP_data = "SP_new_covidcases.csv"
Combined_data="Final_Combine_Data.csv"

```

New York data cleaning

```python
# Read the files and store into Pandas Dataframes using directly the read_csv function that converts excel data into the dataframe
NYCovid_df = pd.read_csv(NY_data, parse_dates=['date'],  encoding="ISO-8859-1",low_memory=False)
```

```python
NYCovid_df.head()
```

```python
#Drop columns not required
NYCovid_df=NYCovid_df.drop(["PROBABLE_CASE_COUNT","New york 2020 population","ALL_CASE_COUNT_7DAY_AVG", "BX_CASE_COUNT","BX_PROBABLE_CASE_COUNT","BX_CASE_COUNT_7DAY_AVG","MN_ALL_CASE_COUNT_7DAY_AVG","QN_CASE_COUNT","QN_PROBABLE_CASE_COUNT","QN_CASE_COUNT_7DAY_AVG","QN_ALL_CASE_COUNT_7DAY_AVG","SI_CASE_COUNT","SI_PROBABLE_CASE_COUNT","SI_CASE_COUNT_7DAY_AVG","SI_ALL_CASE_COUNT_7DAY_AVG","INCOMPLETE", "BX_ALL_CASE_COUNT_7DAY_AVG","BK_CASE_COUNT","BK_PROBABLE_CASE_COUNT","BK_CASE_COUNT_7DAY_AVG","BK_ALL_CASE_COUNT_7DAY_AVG","MN_CASE_COUNT","MN_PROBABLE_CASE_COUNT","MN_CASE_COUNT_7DAY_AVG"], axis=1)
NYCovid_df.head()
```

```python
NYCovid_df.info()
```

```python
#determine any null values in the data
NYCovid_df.isnull().sum()
```

```python
#generate a CSV output of clean file
NYCovid_df.to_csv (r'Clean_CSV_newdata\NYCovid_newdata.csv', index = False, header=True)
```

Sao Paulo Covid file cleanup

```python
SPCovid_df = pd.read_csv(SP_data,parse_dates=['date'], encoding="ISO-8859-1",low_memory=False)
SPCovid_df.head()
```

```python
#Drop columns not required
SPCovid_df=SPCovid_df.drop(["epi_week","country","state","city","code region id","name region","new deaths","deaths","Sao Paulo population","total cases","deaths_per_100K","totalcases_per_100_k","deaths_by_total_cases","data source","last info date","data source for analysis","rec insert dt", "id", "original from wcota git hub"], axis=1)
SPCovid_df.head()
```

```python
#determine any null values in the data
SPCovid_df.isnull().sum()
```

```python
SPCovid_df = SPCovid_df.fillna(0)
```

```python
SPCovid_df.info()
```

```python
#generate a CSV output of clean file
SPCovid_df.to_csv (r'Clean_CSV_newdata\SPCovid_newdata.csv', index = False, header=True)
```
